## Import des librairies

```python
import cv2
import matplotlib.pyplot as plt
```

## Exercice 1

**On charge notre image**

```python
image = cv2.imread('gema.jpg')
```

**On convertit notre image de BGR (OpenCV) en RGB (matplotlib)**

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

**On affiche l'image**

```python
plt.imshow(image)
plt.axis('off')
plt.title('Image GEMA')
plt.show()
```

**On sÃ©pare les canneaux de couleur**

```python
red = image[:, :, 0]
green = image[:, :, 1]
blue = image[:, :, 2]
```

**On calcule les dÃ©rivÃ©es par rapport Ã  x et y**

```python
dx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
dy = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
```

**On affiche notre image originale suivi des 3 image selon les couleurs**

```python
plt.figure(figsize=(15, 10))

plt.subplot(1, 4, 1)
plt.imshow(image)
plt.title('Image originale')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(red)
plt.title('Image rouge')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(green)
plt.title('Image verte')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(blue)
plt.title('Image bleue')
plt.axis('off')

plt.show()
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(1, 2, 1)
plt.imshow(dx, cmap='gray')
plt.title('Image dÃ©rivÃ©e par rapport Ã  x')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(dy, cmap='gray')
plt.title('Image dÃ©rivÃ©e par rapport Ã  y')
plt.axis('off')

plt.show()
```

**On afiche les 3 logos sÃ©parÃ©es**

```python
# On crÃ©e notre fonction qui va zoomer sur notre image
def crop_img(img, x, y, w, h):
    return img[y:y+h, x:x+w]
```

```python
# On affiche l'image avec les axes
plt.imshow(image)
plt.show()
```

```python
# On affiche les 3 logos sÃ©parÃ©s

width = 200
height = 200

img_iaschool = crop_img(image, 200, 400, width, height)
img_esi = crop_img(image, 500, 400, width, height)
img_cyber = crop_img(image, 800, 400, width, height)

plt.figure(figsize=(15, 10))

plt.subplot(1, 3, 1)
plt.imshow(img_iaschool)
plt.title('Image IA School')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_esi)
plt.title('Image ESI')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_cyber)
plt.title('Image Cyber')
plt.axis('off')

plt.show()
```

## Exercice 2

**On convertit nos images en niveau de gris**

```python
img1 = cv2.imread('img1.png')
img2 = cv2.imread('img2.png')

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
fig, ax = plt.subplots()
ax.imshow(img1, cmap='gray')
ax.set_title('Image 1 en niveaux de gris')
ax.axis('off')

# on crÃ©e un rectangle au centre de l'image
height, width = img1.shape
center_x, center_y = width // 2, height // 2
rect_width, rect_height = 50, 100
top_left_x, top_left_y = center_x - rect_width // 2, center_y - rect_height // 2

# On affiche le rectangle
rect = plt.Rectangle((top_left_x, top_left_y), rect_width, rect_height, linewidth=2, edgecolor='r', facecolor='none')
ax.add_patch(rect)
plt.show()
```

**On affiche les diffÃ©rence de nos images**

```python
difference = cv2.absdiff(img1, img2)

plt.figure(figsize=(6, 6))
plt.imshow(difference, cmap='gray')
plt.title('DiffÃ©rence absolue entre les deux images')
plt.axis('off')
plt.show()
```

**On enregistre l'image contenant les diffÃ©rences**

```python
cv2.imwrite('difference.png', difference)
```

**On fait pareil pour les images `maison1.png` et `maison2.png`**
